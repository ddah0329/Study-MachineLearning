```python
import pandas as pd

data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'

df_data = pd.read_csv(data_url, sep='\s+', header = None)
df_data.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'ITAXI', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

df_data.head()
```

```python
import numpy as np

df_data['weight_0'] = 1 # weight_0이라는 이름에 값 1을 넣어서 데이터테이블에 추가

df_data = df_data.drop("MEDV", axis = 1) # MEDV 이름의 Y축 제거

df_data.head()
```

```python
df_matrix = df_data.values # 기존 판다스 객체 데이터를 넘파이 데이터로 변환
```

```python
weight_vector = np.random.random_sample((14,1)) # 14x1 형태로 가중치 값 w 생성
weight_vector
```

```python
df_matrix.dot(weight_vector) # 가중치 벡터와 전체 데이터 X간의 벡터의 내적 연삭 수행
```
